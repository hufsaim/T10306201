```python
import pandas as pd
import numpy as np
```

```python
d = pd.read_csv('20231121.csv')
#d
```

```python
import matplotlib.pyplot as plt
```

```python
X = d.iloc[:,1:4].to_numpy()
y = d.iloc[:,-1].to_numpy()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
for n in range(len(X)):
  ax.scatter(X[n,0],X[n,1],X[n,2],marker='o',c='r')
ax.set_xlabel('feature 1')
ax.set_ylabel('feature 2')
ax.set_zlabel('feature 3')
plt.show()
```

```python
y_hat = np.zeros_like(y)
w = np.array([-1,1,1])
b = np.array([-1])
for n in range(len(X)):
  y_hat[n] = np.sign(np.dot(w,X[n,:]) + b)
```

## display results

```python
fig = plt.figure()
correct = 0
ax = fig.add_subplot(projection='3d')
for n in range(len(X)):
  if y[n]==1 and y_hat[n]==1:
    ax.scatter(X[n,0],X[n,1],X[n,2],marker='o',c='g')
    correct+=1
  if y[n]==1 and y_hat[n]==-1:
    ax.scatter(X[n,0],X[n,1],X[n,2],marker='x',c='g')
  if y[n]==-1 and y_hat[n]==1:
    ax.scatter(X[n,0],X[n,1],X[n,2],marker='x',c='b')
  if y[n]==-1 and y_hat[n]==-1:
    ax.scatter(X[n,0],X[n,1],X[n,2],marker='o',c='b')
    correct+=1


ax.set_xlabel('feature 1')
ax.set_ylabel('feature 2')
ax.set_zlabel('feature 3')
plt.show()

print(f'accuracy: {correct/len(X)}')
```
