```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
```

```python
h = nib.load('ct.nii.gz')
print(h)
```

```python
img = h.get_fdata()
voxelsize = h.header['pixdim'][1:4]
print(voxelsize, img.shape)
```

## Axial

```python
plt.imshow(img[:,:,10],cmap='gray',vmin=-1000,vmax=500)
plt.title('wrong display')
plt.show()
```

```python
plt.imshow(img[:,:,10].transpose((1,0)),cmap='gray',vmin=-1000,vmax=500)
plt.title('correct display')
plt.show()
```

## Coronal

```python
plt.imshow(img[:,210,:],cmap='gray',vmin=-1000,vmax=500)
plt.title('wrong display')
plt.show()
```

```python
plt.imshow(np.rot90(img[:,210,:]),cmap='gray',vmin=-1000,vmax=500)
plt.title('correct?')
plt.show()
```

```python
plt.imshow(np.rot90(img[:,210,:]),cmap='gray',vmin=-1000,vmax=500, aspect=voxelsize[2]/voxelsize[1])
plt.title('correct display')
plt.show()
```
