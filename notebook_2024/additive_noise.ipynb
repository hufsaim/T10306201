```python
f = plt.imread(imgpath+'/test-pattern.tif')
print(f.shape,f.dtype)
f = f.astype(np.int16)
print(f.shape,f.dtype)
plt.imshow(f,cmap='gray',vmin=0,vmax=255)
plt.show()
```

```python
x = np.arange(0,256)
dist = np.exp(-(x)/(32))
dist = dist/np.sum(dist)
plt.bar(x,dist)
plt.show()
```

```python
g = np.zeros_like(f)
for i in np.arange(f.shape[0]):
  for j in np.arange(f.shape[1]):
    noise = np.random.choice(x,p=dist)
    g[i,j] = f[i,j] + noise
plt.imshow(g,cmap='gray',vmin=0,vmax=255)
plt.show()
```
