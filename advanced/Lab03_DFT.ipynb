```python
import numpy as np
import matplotlib.pyplot as plt
```

### DFT matrix for N = 8

```python
N = 8
n = np.arange(N)
E = np.zeros((N,N)).astype(complex)
for k in range(N):
    E[k] = np.exp(1j*2*np.pi*k*n/N)

plt.subplot(121)
plt.imshow(np.real(E),cmap='gray')
plt.title('real')
plt.subplot(122)
plt.imshow(np.imag(E),cmap='gray')
plt.title('imaginary')
plt.show()
```

### plot basis functions

```python
for k in range(N):
    plt.subplot(121)
    plt.stem(np.real(E[k]))
    plt.axis([-1,N,-1.2,1.2])
    plt.title(f'basis {k}, real')

    plt.subplot(122)
    plt.stem(np.imag(E[k]))
    plt.axis([-1,N,-1.2,1.2])
    plt.title(f'basis {k}, imaginary')

    plt.show()
```

### Test with an arbitrary sequence

```python
x = np.array([3,6,9,3,6,9,0,0])
plt.stem(x)
plt.show()
```

### Calculate Fourier Coefficients

```python
X = np.dot(np.conj(E),x)
X
```

### Check synthetic equation

```python
x_recon = np.zeros_like(x).astype(complex)
for k in range(N):
    x_recon += X[k]*E[k]
x_recon = (1/N)*x_recon
```

```python
plt.stem(np.real(x_recon))
plt.show()
```

- the same result can be calculated by Matrix-Vector Multiplication

```python
plt.stem(np.real((1/N)*np.dot(E,X)))
plt.show()
```
