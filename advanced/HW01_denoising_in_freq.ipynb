```python
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np
```

```python
d = sio.loadmat('noisydata.mat')
F1 = d['F1']
F2 = d['F2']

print(F1.shape)
print(F2.shape)
```

```python
img1 = np.fft.fftshift(np.fft.fftn(np.fft.fftshift(F1)))
plt.figure(dpi=250)
plt.subplot(211)
plt.imshow(np.log(np.abs(F1)),cmap='gray')
plt.subplot(212)
plt.imshow(np.abs(img1),cmap='gray',vmin=0,vmax=1.2)
plt.tight_layout()
plt.show()
```
