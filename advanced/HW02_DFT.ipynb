```python
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np
```

```python
d = sio.loadmat('kspace_2d_fft_5echo.mat')
kspace = d['kspace']
te = d['te']
print(kspace.shape) # x, y, z, t
print(te) # sec
```

```python
N = 32
Nc = int(N/2)
n = np.arange(-Nc,Nc)
A = np.zeros((N,N)).astype(complex)
for k in range(N):
    A[k] = np.exp(1j*2*np.pi*(k-(Nc))*n/N)

At = np.matrix.conjugate(A)
plt.subplot(221)
plt.imshow(np.real(A))
plt.title('A_real')
plt.subplot(222)
plt.imshow(np.real(At))
plt.title('At_real')
plt.subplot(223)
plt.imshow(np.imag(A))
plt.title('A_imag')
plt.subplot(224)
plt.imshow(np.imag(At))
plt.title('At_imag')
plt.tight_layout()
plt.show()
```

```python
F32 = kspace[160-16:160+16,160-16:160+16,0,2]
F32.shape
```

```python
plt.subplot(141)
plt.imshow(np.abs(np.dot(A,F32)),cmap='gray')
plt.title('(1)')

plt.subplot(142)
plt.imshow(np.abs(np.dot(F32,At)),cmap='gray')
plt.title('(2)')

plt.subplot(143)
plt.imshow(np.abs(np.dot(A,np.dot(F32,At))),cmap='gray')
plt.title('(3)')

plt.subplot(144)
plt.imshow(np.abs(np.dot(At,np.dot(F32,A))),cmap='gray')
plt.title('(4)')

plt.tight_layout()

plt.show()
```

## Q2. 아래와 같이 정의된 Fourier domain data에 대해 Fourier transform matrix를 정의하고, Fourier transform matrix를 활용하여 복원한 결과에 대해 magnitude, phase, real, imag를 각각 출력

```python
F = kspace[:,:,0,2]
F.shape

plt.subplot(121)
plt.imshow(np.log(np.abs(F)))
plt.title('Magnitude in Fourier spectrum')
plt.subplot(122)
plt.imshow(np.angle(F))
plt.title('Phase in Fourier spectrum')
plt.tight_layout()
plt.show()
```
