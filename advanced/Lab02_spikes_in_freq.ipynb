```python
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np
```

```python
d = sio.loadmat('kspace_2d_fft_5echo.mat')
kspace = d['kspace']
te = d['te']

print(kspace.shape) # x, y, z, t
print(te) # sec
```

```python
F = kspace[:,:,0,2]
F.shape
```

```python
img = np.fft.fftshift(np.fft.fftn(np.fft.fftshift(F)))
plt.subplot(121)
plt.imshow(np.log(np.abs(F)),cmap='gray')
plt.subplot(122)
plt.imshow(np.abs(img),cmap='gray')
plt.show()
```

```python
eps=1e-3
H = np.zeros(F.shape)
u0 = 160
v0 = 160
F_x = F.copy()
F_x[u0-1:u0+1,v0-1:v0+1]=5*np.max(np.abs(F))
img_x = np.fft.fftshift(np.fft.fftn(np.fft.fftshift(F_x)))
plt.subplot(131)
plt.imshow(np.log(np.abs(F_x)+eps),cmap='gray')
plt.axis('off')
plt.subplot(132)
plt.imshow(np.abs(img),cmap='gray')
plt.axis('off')
plt.subplot(133)
plt.imshow(np.abs(img_x),cmap='gray')
plt.axis('off')
plt.show()
```
