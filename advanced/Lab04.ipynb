```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
d = sio.loadmat('kspace_2d_fft_5echo.mat')
```

```python
kspace = d['kspace']
te = d['te']
print(te) # echo time (ms)
kspace.shape,te.shape
```

```python
Ns = 7
Ne = 4
F = kspace[:,:,Ns,Ne]
M = np.fft.fftshift(np.fft.fftn(np.fft.fftshift(F)))
```

## Full sampling (Nyquist)

```python
plt.figure(dpi=150)
plt.subplot(121)
plt.imshow(np.abs(M),cmap='gray')
plt.axis('off')
plt.subplot(122)
plt.imshow(np.angle(M),cmap='gray')
plt.axis('off')
plt.show()
```

## 2x undersampling (y-direction)

```python
Us = np.zeros_like(F)
Us[::2,:] = 1
#Us[160-16:160+16,:]=1
Mu = np.fft.fftshift(np.fft.fftn(np.fft.fftshift(F*Us)))
```

```python
plt.figure(dpi=150)
plt.subplot(121)
plt.imshow(np.abs(Mu),cmap='gray')
plt.axis('off')
plt.subplot(122)
plt.imshow(np.angle(Mu),cmap='gray')
plt.axis('off')
plt.show()
```

## 2x random undersampling (y-direction)

```python
Us = np.zeros_like(F)
Ry = np.array(random.sample(list(range(Us.shape[-1])),int(0.5*Us.shape[-1])))
Us[Ry,:] = 1
#Us[160-16:160+16,:]=1
Mu = np.fft.fftshift(np.fft.fftn(np.fft.fftshift(F*Us)))
```
